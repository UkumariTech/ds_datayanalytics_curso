# **Ejercicio N°1:**
Escribir un programa que pregunte al usuario o genere las ventas
de los últimos 5 días, y muestre por pantalla una serie (Dataframe)
con los datos de las ventas por producto (opcional): Nro de
unidades y precio unitario, indexada por los días, antes y después
de aplicarles un descuento del 5%.
(Diversas opciones de solución: libre)

```python
dias=['Día 1','Día 2','Día 3','Día 4','Día 5'] # lista con los 5 días de ventas
#Se genera con ayuda de un for la cantidad de ventas en esos 5 días
ventas=[]
for i in range(0,5):
  print(f'Ingrese las ventas del día {i+1}')
  a=int(input(' \n'))
  ventas.append(a)
print(ventas)
```

```python
# Se almacena los datos en un diccionario
dict_ventas={'Días': dias, 'Cantidad':ventas}
print('-----Diccionario-----')
print(dict_ventas)
# se crea dataframe
import pandas as pd
df_ventas=pd.DataFrame(dict_ventas)#, index='Días')
print('--------DATAFRAME--------')
print(df_ventas)
```

```python
import random
PU_random=[]
for i in range(0,5):
  b_ran=random.randint(21,65)
  PU_random.append(b_ran)
print(PU_random)
#se agrega al dataframe
df_ventas['Precio Unitario']=PU_random
print(df_ventas)
```

```python
#calcular el total de ventas por fila
def TotalVentasPorDia(fila):
  respuesta=fila['Cantidad'] * fila['Precio Unitario']
  return respuesta
# Se agrega en una nueva columna la multiplicacion de la venta por el precio unitario
df_ventas['Total Ventas'] = df_ventas.apply(TotalVentasPorDia, axis=1) # al usar 1 se indica que la operacion es por cada fila
print(df_ventas)
```

```python
# se agrega la columna con el descuento del 5%
def Descuento(TotalVentas):
  descuento = TotalVentas * 0.95
  return descuento

df_ventas['Descuento 5%'] = df_ventas['Total Ventas'].apply(Descuento)
print(df_ventas)
```
