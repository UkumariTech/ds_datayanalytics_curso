# **Ejercicio N°4:**
Con el siguiente fichero datos_seguros.csv contiene información de riesgo de clientes de un banco. Escribir un programa con los siguientes requisitos:
1. Generar un DataFrame con los datos del fichero.
2. Mostrar por pantalla las dimensiones del DataFrame, el número de datos que contiene, los nombres de sus columnas y filas, los tipos de datos de las columnas, mostrar las 10 primeras filas y las 10 últimas filas, verificar si tiene datos perdidos/nulos.
3. Extraer las 1000 primeras filas y almacenarlas en un contenedor.

```python
import pandas as pd
'''
1 Generar un DataFrame con los datos del fichero.
'''
#leer csv
df = pd.read_csv("Data_Riesgos.csv", index_col=['ID'])
print(df.head())
```

```python
'''
2. Mostrar por pantalla las dimensiones del DataFrame, el número de datos que contiene, los nombres de sus columnas y filas,
los tipos de datos de las columnas, mostrar las 10 primeras filas y las 10 últimas filas, verificar si tiene datos perdidos/nulos.
'''
#primero se obtiene una informacion global del df para ir comprobando con los pasos que vienen
print('---informacion df--- \n')
print(df.info())
# dimensiones del data frame
print('---------------------------------------------------------------------------------------------------------------------------------')
print('*******1 Dimensiones del data frame******* \n')
print('Las dimensiones del DATAFRAME es: ', df.shape , '->(filas, columnas) \n')

# Número de datos que contiene
print('---------------------------------------------------------------------------------------------------------------------------------')
print('*******2 Número de datos que contiene******* \n')
print('El tamaño del DATAFRAME es: ', df.size, 'datos \n') # devuelve filas * columnas

# Nombres de las columnas y filas
print('---------------------------------------------------------------------------------------------------------------------------------')
print('*******3 Nombres de las columnas y filas******* \n')
columnas_nombres = df.columns.values
filas_nombres = df.index.values
print('Nombres de las columnas: ', columnas_nombres, '\n')
print('Nombres de las filas: ', filas_nombres, '\n') # se muestran pocos ya que hay 4500 filas

#Tipos de datos de las columnas
print('---------------------------------------------------------------------------------------------------------------------------------')
print('*******4 Tipos de datos de las columnas******* \n')
#print(df.columns['edad'].type())
df_types = df.dtypes
print(df_types)


# Mostrar las 10 primeras filas
print('---------------------------------------------------------------------------------------------------------------------------------')
print('*******5 Mostrar las 10 primeras filas******* \n')
print(df[0:10])
# Mostrar las 10 últimas filas
print('---------------------------------------------------------------------------------------------------------------------------------')
print('*******6 Mostrar las 10 últimas filas******* \n')
print(df.iloc[4490:4500]) # filtrado de filas por índices

#Verificar si tiene datos perdidos/nulos
print('---------------------------------------------------------------------------------------------------------------------------------')
print('*******7 Verificar si tiene datos perdidos/nulos******* \n')
print(df.isnull().sum())
```

```python
#3. Extraer las 1000 primeras filas y almacenarlas en un contenedor.
contenedor = df.head(1000)
print(contenedor, '\n')
print('-----info -----\n')
print(contenedor.info())
```
