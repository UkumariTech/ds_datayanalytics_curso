# **Ejercicio N°2:**

Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa, y devuelva una serie con el ingreso mínimo, máximo y media.

El resultado final debe ser una serie con los estadísticos.


```python
import pandas as pd

def estadistica_notas(mesesIng):
    meses = pd.Series(mesesIng)
    std = pd.Series([meses.min(), meses.max(), meses.mean()], index=['Min', 'Max', 'Media'])
    return std

mesesIng = {'ENERO':9, 'FEBRERO':6.5, 'MARZO':4, 'ABRIL': 8.5, 'MAYO': 5, 'JUNIO':7,'JULIO':11, 'AGOSTO':54.5, 'SEP':42, 'OCTUBRE': 18.5, 'NOV': 5, 'DIC':7}
print(mesesIng )
print(estadistica_notas(mesesIng ))
```
