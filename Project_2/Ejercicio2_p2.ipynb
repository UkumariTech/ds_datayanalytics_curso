- Generar 5 nuevas variables aplicando los siguientes criterios con las
variables numéricas:
  1. Log10(var1/var2)
  2. Sqrt(var1)*exp(var2)/200
  3. Si var 1 > var 2 → 5, caso contrario 3
  4. 1/logn(var1/var2)*100
  5. var2**2/var1

```python
import math
def col_uno(fila):
  # columnas: Desplazamiento y Aceleracion
  log10 = math.log((fila['Desplazamiento'] / fila['Aceleracion']),10)

  return log10

df['Log10(var1/var2)'] = df.apply(col_uno, axis=1)
df.head()
```

```python
# Sqrt(var1)*exp(var2)/200

def col_dos(fila):
  # columnas: millas y peso
  #log10 = math.log((fila['Desplazamiento'] / fila['Aceleracion']),10)
  raiz2 = np.sqrt(fila['Millas_Por_Galon'])
  exponente = math.pow(fila['Peso_En_Libras'] , 2) # al cuadrado
  division = exponente / 200
  resp_final = raiz2 * division
  return resp_final

df['Sqrt(var1)*exp(var2)/200'] = df.apply(col_dos, axis=1)
df.head()
```

```python
# Si var 1 > var 2 → 5, caso contrario 3

def reemplazo(fila):
  # 	var2: Millas_Por_Galon y var1: Aceleracion
  if fila['Aceleracion'] > fila['Millas_Por_Galon']:
    return 5
  else:
    return 3

df['5 o 3'] = df.apply(reemplazo, axis=1)
df.head()
```

```python
# 1/logn(var1/var2)*100

def col_cuatro(fila):
  # 5 o 3 y Aceleracion
  division1 = ((fila['Aceleracion'])/(fila['5 o 3']))
  Ln = math.log(division1, np.e)
  mult = (Ln * 100)
  division2 = ((1) / (mult))

  return division2

df['1/logn(var1/var2)*100'] = df.apply(col_cuatro, axis=1)
df.head()
```

```python
# var2**2/var1
def col_cinco(fila):
  # cilindros y pesos en libras

  expon = math.pow(fila['Cilindros'] ,2)
  div = ((expon) / (fila['Peso_En_Libras']))
  return div

df['var2**2/var1'] = df.apply(col_cinco, axis=1)
df.head()
```
