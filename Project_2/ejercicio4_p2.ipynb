```python
#importar librerías

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/Curso_DataScience_peru/clases/13.2_PROYECTO2_27JULIO2023/autos_RespEjer3.csv',delimiter=",")
df.head()
```

```python
# extraer columnas Caballo de fuerza y Variables cualitativas
df2 = df[['Caballo_De_Fuerza','Variables_Cualitativas']]
df2
```

```python
def graficas2(Variables_Cualitativas):

  '''
  Si dato columna Caballo_De_Fuerza es igual a - Alto - de la columna Variables Cualitativas se genera gráfica - Cajas y bigotes -
  Si dato columna Caballo_De_Fuerza es igual a - Medio - de la columna Variables Cualitativas se genera gráfica - Barras de la media -
  Si dato columna Caballo_De_Fuerza es igual a - Bajo - de la columna Variables Cualitativas se genera gráfica - Histograma con densidad (Curva) -
  '''
  df_alto = df2.copy()
  df_alto.drop(df_alto.loc[df_alto['Variables_Cualitativas']=='Medio'].index, inplace=True)
  df_alto.drop(df_alto.loc[df_alto['Variables_Cualitativas']=='Bajo'].index , inplace=True)
  df_alto_ordenado= df_alto.sort_values('Caballo_De_Fuerza') # se debe ordenar de mayor a menor
  plt.subplot(1,3,1)
  plt.title('Caja y Bigote')
  plt.boxplot(df_alto_ordenado['Caballo_De_Fuerza'], vert=False)

  df_medio = df2.copy()
  df_medio.drop(df_medio.loc[df_medio['Variables_Cualitativas']=='Alto'].index, inplace=True)
  df_medio.drop(df_medio.loc[df_medio['Variables_Cualitativas']=='Bajo'].index , inplace=True)
  plt.subplot(1,3,2)
  plt.title('Barras de la media')
  sns.barplot(estimator=np.mean, data=df_medio['Caballo_De_Fuerza'])

  df_bajo = df2.copy()
  df_bajo.drop(df_bajo.loc[df_bajo['Variables_Cualitativas']=='Alto'].index, inplace=True)
  df_bajo.drop(df_bajo.loc[df_bajo['Variables_Cualitativas']=='Medio'].index , inplace=True)
  plt.subplot(1,3,3)
  plt.title('Histograma con densidad')
  #df_bajo['Caballo_De_Fuerza'].plot(kind='density',linestyle='--', color='r')
  sns.displot(data=df_bajo['Caballo_De_Fuerza'], kind='kde')

```
