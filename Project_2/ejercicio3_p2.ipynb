Crear una función que permita generar 3 variables cualitativas
ordinales a partir de las variables numéricas con los siguientes criterios:



*   Si es mayor a la “media + 1*desviación estándar” → Alto
*   Si está entre “media - 1*desviación estándar” y “media +
    1*desviación estándar” → Medio
*   Si es menor a la “media - 1*desviación estándar” → Bajo

Nota: Mostrar con las nuevas variables generadas.

Elegir con libertad las variables a trabajar.

La función debe contar con su respectiva documentación.

```python
#importar librerías

import pandas as pd
import numpy as np
df = pd.read_csv('/content/drive/MyDrive/Curso_DataScience_peru/clases/13.2_PROYECTO2_27JULIO2023/autos.csv',delimiter=";")
df.head()
```

```python
#Verificar si tiene datos perdidos/nulos
print('******* Datos perdidos/nulos ******* \n')
print(df.isnull().sum())
```

```python
# Se saca una DF con los estadísticos

df.describe()
```

```python
mean_Millas_Por_Galon = df['Millas_Por_Galon'].mean()
mean_Millas_Por_Galon = round(mean_Millas_Por_Galon,2)
print(mean_Millas_Por_Galon)

mean_Caballo_De_Fuerza = df['Caballo_De_Fuerza'].mean()
mean_Caballo_De_Fuerza = round(mean_Caballo_De_Fuerza,2)
print(mean_Caballo_De_Fuerza)
```

```python
#Cambiar datos NaN por el promedio o media de la columna

df['Millas_Por_Galon'] = df['Millas_Por_Galon'].replace(np.nan, mean_Millas_Por_Galon )
df['Caballo_De_Fuerza'] = df['Caballo_De_Fuerza'].replace(np.nan, mean_Caballo_De_Fuerza)

#Verificar si tiene datos perdidos/nulos
print('******* Datos perdidos/nulos ******* \n')
print(df.isnull().sum())
```

```python
def variables_cualitativas(Caballo_De_Fuerza):
  '''
  1   Si es mayor a la “media + 1*desviación estándar” → Alto
  2   Si está entre “media - 1*desviación estándar” y “media +
      1*desviación estándar” → Medio
  3   Si es menor a la “media - 1*desviación estándar” → Bajo
  '''
  mean = 105.082500
  std = 38.768779

  if Caballo_De_Fuerza > (mean + 1*std):
    return 'Alto'
  elif (mean -1*std) <= Caballo_De_Fuerza <= (mean + 1*std):
    return 'Medio'
  else:
    return 'Bajo'
```

```python
df['Variables_Cualitativas'] = df['Caballo_De_Fuerza'].apply(variables_cualitativas)
df.head(45)
```
