Generar una función y utilizar esta para recodificar la variable “país de
origen” (1, 2, 3) en el dataframe a través de la función apply.

```python
#importar librerías

import pandas as pd
#import random
#import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Curso_DataScience_peru/clases/13.2_PROYECTO2_27JULIO2023/autos.csv',delimiter=";")
df.head(22)

```

```python
print('---Info---')
df.info()
```

```python
# primero se crea la función

def recodificar(Pais_origen):
  if Pais_origen == 'USA':
    return 1
  elif Pais_origen == 'Europe':
    return 2
  else:
    return 3
```

```python
# se aplica la función

df['Pais_Origen']  = df['Pais_Origen'].apply(recodificar)
df.head(22)
```
