```python
#importar librerías

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('/content/drive/MyDrive/Curso_DataScience_peru/clases/13.2_PROYECTO2_27JULIO2023/autos_RespEjer3.csv',delimiter=",")
df.head()
```

```python
df.describe()
```

```python
# extraer columnas Caballo de fuerza y Variables cualitativas
df2 = df[['Caballo_De_Fuerza','Variables_Cualitativas']]
df2
```

```python
df2.describe()
```

```python
# eliminar filas con Medio y Bajo

                              # CAJA Y BIGOTES

df_alto = df2.copy()
df_alto.drop(df_alto.loc[df_alto['Variables_Cualitativas']=='Medio'].index, inplace=True)
df_alto.drop(df_alto.loc[df_alto['Variables_Cualitativas']=='Bajo'].index , inplace=True)
print(df_alto.head())
print('--Caja y Bigote-- \n')
df_alto_ordenado= df_alto.sort_values('Caballo_De_Fuerza') # se debe ordenar de mayor a menor
print(df_alto_ordenado)
print('--Caja y bigotes-- \n')
plt.boxplot(df_alto_ordenado['Caballo_De_Fuerza'], vert=False)
plt.grid(True)
plt.show()
```

```python
# eliminar filas con Alto y Bajo

                          # Barras de la media

df_medio = df2.copy()
df_medio.drop(df_medio.loc[df_medio['Variables_Cualitativas']=='Alto'].index, inplace=True)
df_medio.drop(df_medio.loc[df_medio['Variables_Cualitativas']=='Bajo'].index , inplace=True)
print(df_medio.head())
print('--Descripcion-- \n')
print(df_medio.describe())
print('--Barras de la media-- \n')
import seaborn as sns
sns.barplot(estimator=np.mean,
            data=df_medio)
plt.show()

# https://www.geeksforgeeks.org/barplot-using-seaborn-in-python/
```

```python
# eliminar filas con Alto y Medio
                  # Histograma de densidad

df_bajo = df2.copy()
df_bajo.drop(df_bajo.loc[df_bajo['Variables_Cualitativas']=='Alto'].index, inplace=True)
df_bajo.drop(df_bajo.loc[df_bajo['Variables_Cualitativas']=='Medio'].index , inplace=True)
print(df_bajo.head())
#print('--Descripcion-- \n')
#print(df_bajo.describe())
print('--Histograma-- \n')
plt.hist(df_bajo['Caballo_De_Fuerza'], ec='black')
plt.grid(True)
plt.legend()
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Caballos de fuerza')
plt.show()
print('--Histograma de densidad-- \n')
plt.figure()
df_bajo['Caballo_De_Fuerza'].plot(kind='density',linestyle='--', color='r')
plt.grid(True)
plt.show()
```
