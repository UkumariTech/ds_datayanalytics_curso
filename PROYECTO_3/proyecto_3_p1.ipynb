# Generar nuevas variables de valor

image.png

```python
def IMC1(IMC):
  if IMC < 16:
    return 'Delgadez_servera'
  elif 16<=IMC<17:
    return 'Delgadez_moderada'
  elif 17<=IMC<18.5:
    return 'Delgadez_leve'
  elif 18.5<=IMC<24.5:
    return 'Normal'
  elif 24.5<=IMC<30:
    return 'Sobrepeso'
  elif 30<=IMC<35:
    return 'Obesidad_tipo_1'
  elif 35<=IMC<40:
    return 'Obesidad_tipo_1'
  else:
    return 'Obesidad_tipo_3'
```

```python
df2['lec_IMC']=df2['IMC'].apply(IMC1)
df2.head()
```

```python
names_columns_reindex=df2.columns
names_columns_reindex
```

```python
df2 = df2.reindex(columns=['talla', 'edad', 'peso_kg', 'circun_cuello', 'IMC','lec_IMC', 'circun_cintura',
       'cadera', 'ind_cintura_cadera', 'ind_cintura_estatura',
       'por_grasa_corporal', 'masa_corporal_magra_kg',
       'pliegue_cutaneo_BICEPS', 'pliegue_cutaneo_TRICEPS',
       'pliegue_cutaneo_ESCAPULAR', 'pliegue_cutaneo_SUPRAILIACO'])
df2.head()
```
